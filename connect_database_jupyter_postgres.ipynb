Substituir variáveis user e pass pelos respetivos username e password.

```python
import os

user = os.getenv('user')
password = os.getenv('pass')
connection_string = "postgresql://postgres:pass@localhost/MIEBIOM".format(user=user, password=password)
%sql $connection_string
```

```python
%%sql postgresql://postgres:pass@localhost/MIEBIOM
select freguesia from braga
```

```python
%%sql postgresql://postgres:pass@localhost/MIEBIOM
select freguesia, concelho, distrito from portugal
```
