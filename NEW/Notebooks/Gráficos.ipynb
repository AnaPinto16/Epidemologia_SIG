```python
#Importar as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16,7)

import warnings
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
import numpy as np; np.random.seed(1)
import seaborn as sns
sns.set(style="ticks", rc={"lines.linewidth": 1.6})
sns.set_palette('bright')
from datetime import date
import io
import os, math

import matplotlib.animation as animation
from IPython.display import HTML
```

```python
base = pd.read_csv('owid-covid-data.csv')
base
```

# Gráfico com o Total de casos, Total de novos casos e Total de mortes

```python
base.groupby(['date'])['total_cases'].sum().plot()
base.groupby(['date'])['new_cases'].sum().plot()
base.groupby(['date'])['total_deaths'].sum().plot()
base.groupby(['date'])['new_deaths'].sum().plot()
plt.title("Total de casos, novos casos, número total de Mortes e novos casos")
plt.xlabel("Data")
plt.ylabel("Número de casos")
plt.legend()
plt.grid(True)
```

# Gráfico Crescimento do Nº de casos em Itália em maio

```python
casos_df = pd.read_csv('ITA2.csv')
datas = casos_df['date']
datas = ['/'.join(d.split('/')[::-1]) for d in datas]
casos_df['date'] = np.asarray(datas)
casos_df.drop(0, inplace=True)
display(casos_df)
```

```python
dados = casos_df[['date', 'total_cases', 'total_cases']]
dados = dados.melt('date', var_name='Casos', value_name='total_cases')
display(dados)
```

```python
plt.figure(figsize=(20, 5))

sns.barplot(data=dados, x='date', y='total_cases', hue='Casos', dodge=False, palette='husl')
plt.title('Gráfico Crescimento do Nº de casos em Itália em maio')
plt.xticks(rotation=45)
```

```python
base.groupby(['date'])['total_cases'].sum().plot()
plt.title("Total de casos em Itália desde o início da pandemia")
plt.xlabel("Datas")
plt.ylabel("Número de casos")
plt.legend()
plt.grid(True)
```

# Gráfico Crescimento do Nº de mortes em Itália em maio

```python
dados = casos_df[['date', 'total_deaths', 'total_deaths']]
dados = dados.melt('date', var_name='Casos', value_name='total_deaths')
display(dados)
```

```python
plt.figure(figsize=(20, 5))

sns.barplot(data=dados, x='date', y='total_deaths', hue='Casos', dodge=False, palette='husl')
plt.title('Gráfico Crescimento do Nº de mortes em Itália em maio')
plt.xticks(rotation=45)
```

```python
base.groupby(['date'])['total_deaths'].sum().plot()
plt.title("Total de mortes em Itália desde o início da pandemia")
plt.xlabel("Datas")
plt.ylabel("Número de casos")
plt.legend()
plt.grid(True)
```
