# Gráficos comparativos do número de testes realizados em países desenvolvidos e não desenvolvidos

```python
#Importar as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16,7)

import warnings
warnings.filterwarnings('ignore')
import matplotlib.ticker as ticker
import numpy as np; np.random.seed(1)
import seaborn as sns
sns.set(style="ticks", rc={"lines.linewidth": 1.6})
sns.set_palette('bright')
from datetime import date
import io
import os, math

import matplotlib.animation as animation
from IPython.display import HTML
```

```python
base = pd.read_csv('owid-covid-data.csv')
base
```

# Gráfico de avaliação dos testes feitos

```python
base.groupby(['date'])['total_tests'].sum().plot()
base.groupby(['date'])['new_tests'].sum().plot()
base.groupby(['date'])['new_tests_smoothed'].sum().plot()
plt.title("Testes")
plt.xlabel("Data")
plt.ylabel("Número Testes")
plt.legend()
plt.grid(True)
```

# Gráfico de avaliação dos testes feitos Top 5 dos países desenvolvidos

```python
dados_br = base[(base['location'] == 'Brazil')]
dados_it = base[(base['location'] == 'Italy')]
dados_us = base[(base['location'] == 'United States')]
dados_rss = base[(base['location'] == 'Russia')]
dados_es = base[(base['location'] == 'Spain')]
```

```python
dados_it.groupby(['date'])['total_tests'].sum().plot(label='Italia')
dados_br.groupby(['date'])['total_tests'].sum().plot(label='Brazil')
dados_us.groupby(['date'])['total_tests'].sum().plot(label='United States')
dados_rss.groupby(['date'])['total_tests'].sum().plot(label='Russia')
dados_es.groupby(['date'])['total_tests'].sum().plot(label='Spain')
plt.legend()
plt.title('Evolução de Total de Casos no Top5 de Países com COVID-19')
plt.grid(True)
```

# Gráfico de avaliação de novos testes feitos Top 5 dos países desenvolvidos

```python
dados_it.groupby(['date'])['new_tests'].sum().plot(label='Italia')
dados_br.groupby(['date'])['new_tests'].sum().plot(label='Brazil')
dados_us.groupby(['date'])['new_tests'].sum().plot(label='United States')
dados_rss.groupby(['date'])['new_tests'].sum().plot(label='Russia')
dados_es.groupby(['date'])['new_tests'].sum().plot(label='Spain')
plt.legend()
plt.title('Evolução de Total de Casos no Top5 de Países com COVID-19')
plt.grid(True)
```

# Gráfico de avaliação de total de testes feitos Top 5 dos países não desenvolvidos

```python
dados_an = base[(base['location'] == 'Angola')]
dados_tls = base[(base['location'] == 'Timor')]
dados_np = base[(base['location'] == 'Nepal')]
dados_ht = base[(base['location'] == 'Haiti')]
dados_mad = base[(base['location'] == 'Madagascar')]
```

```python
dados_an.groupby(['date'])['total_tests'].sum().plot(label='Angola')
dados_tls.groupby(['date'])['total_tests'].sum().plot(label='Timor')
dados_np.groupby(['date'])['total_tests'].sum().plot(label='Nepal')
dados_ht.groupby(['date'])['total_tests'].sum().plot(label='Haiti')
dados_mad.groupby(['date'])['total_tests'].sum().plot(label='Madagascar')
plt.legend()
plt.title('Evolução de Total de Casos no Top5 de Países com COVID-19')
plt.grid(True)
```

# Gráfico de avaliação de novos testes feitos Top 5 dos países não desenvolvidos

```python
dados_an.groupby(['date'])['new_tests'].sum().plot(label='Angola')
dados_tls.groupby(['date'])['new_tests'].sum().plot(label='Timor')
dados_np.groupby(['date'])['new_tests'].sum().plot(label='Nepal')
dados_ht.groupby(['date'])['new_tests'].sum().plot(label='Haiti')
dados_mad.groupby(['date'])['new_tests'].sum().plot(label='Madagascar')
plt.legend()
plt.title('Evolução de Total de Casos no Top5 de Países com COVID-19')
plt.grid(True)
```

# Gráfico de comparação do nr de testes feitos em Nepal e USA

```python
dados_np = base[(base['location'] == 'Nepal')]
dados_us = base[(base['location'] == 'United States')]
```

```python
dados_np.groupby(['date'])['new_tests'].sum().plot(label='Nepal')
dados_us.groupby(['date'])['new_tests'].sum().plot(label='United States')
plt.legend()
plt.title('Evolução de Total de Casos no Top5 de Países com COVID-19')
plt.grid(True)
```
