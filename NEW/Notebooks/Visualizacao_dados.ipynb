```python
base = pd.read_csv('owid-covid-data.csv')
base
```

Observação do dados divididos por: quantidade, a média, os quartis, desvo padrão, mínimo e máximo.

# Exemplos de Gráficos para facilitar a visualização dos dados

```python
fig = base2.male_smokers.value_counts().plot.bar()
fig.set_title('Nr de Fumadores do sexo masculino')
fig.set_ylabel('Nr de casos')
```

```python
fig2 = base2.female_smokers.value_counts().plot.bar()
fig2.set_title('Nr de Fumadores do sexo feminino')
fig2.set_ylabel('Nr de casos')
```

```python
fig3 = base2.hospital_beds_per_100k.value_counts().plot.bar()
fig3.set_title('Nr de camas por 100k')
fig3.set_ylabel('Nr de casos')
```
